```python
def plot_sig_vs_bkg(
    df_sig,
    df_bkg,
    var,
    cfg,
    tag="Down_mum_tag"
):
    h_sig = ROOT.TH1F(
        f"h_{var}_sig",
        f"{var}; {cfg['xlabel']}; Events",
        cfg["bins"], cfg["xmin"], cfg["xmax"]
    )

    h_bkg = ROOT.TH1F(
        f"h_{var}_bkg",
        f"{var}; {cfg['xlabel']}; Events",
        cfg["bins"], cfg["xmin"], cfg["xmax"]
    )

    for x in df_sig[var]:
        h_sig.Fill(x)

    for x in df_bkg[var]:
        h_bkg.Fill(x)

    # Normalización
    if h_sig.Integral() > 0:
        h_sig.Scale(1.0 / h_sig.Integral())
    if h_bkg.Integral() > 0:
        h_bkg.Scale(1.0 / h_bkg.Integral())

    # Estilo
    h_sig.SetLineColor(ROOT.kRed)
    h_sig.SetLineWidth(2)

    h_bkg.SetLineColor(ROOT.kBlack)
    h_bkg.SetLineWidth(2)

    ymax = 1.2 * max(h_sig.GetMaximum(), h_bkg.GetMaximum())
    h_sig.SetMaximum(ymax)

    c = ROOT.TCanvas(f"c_{var}", var, 800, 600)

    h_sig.Draw("HIST")
    h_bkg.Draw("HIST SAME")

    leg = ROOT.TLegend(0.65, 0.7, 0.88, 0.88)
    leg.AddEntry(h_sig, "Signal (MC)", "l")
    leg.AddEntry(h_bkg, "Background (data)", "l")
    leg.Draw()

    c.SaveAs(f"{var}_{tag}_sig_vs_bkg.png")
```

```python

for var, cfg in var_cfg.items():
    plot_sig_vs_bkg(
        df_sig=MCMumDowntag_sig,
        df_bkg=dataMumDowntag_bkg,
        var=var,
        cfg=var_cfg[var],
        tag="Down_mum_tag"
    )

```

```python
def plot_sig_vs_bkg(
    df_sig,
    df_bkg,
    var,
    cfg,
    tag="VM_mum_tag"
):
    h_sig = ROOT.TH1F(
        f"h_{var}_sig",
        f"{var}; {cfg['xlabel']}; Events",
        cfg["bins"], cfg["xmin"], cfg["xmax"]
    )

    h_bkg = ROOT.TH1F(
        f"h_{var}_bkg",
        f"{var}; {cfg['xlabel']}; Events",
        cfg["bins"], cfg["xmin"], cfg["xmax"]
    )

    for x in df_sig[var]:
        h_sig.Fill(x)

    for x in df_bkg[var]:
        h_bkg.Fill(x)

    # Normalización
    if h_sig.Integral() > 0:
        h_sig.Scale(1.0 / h_sig.Integral())
    if h_bkg.Integral() > 0:
        h_bkg.Scale(1.0 / h_bkg.Integral())

    # Estilo
    h_sig.SetLineColor(ROOT.kRed)
    h_sig.SetLineWidth(2)

    h_bkg.SetLineColor(ROOT.kBlack)
    h_bkg.SetLineWidth(2)

    ymax = 1.2 * max(h_sig.GetMaximum(), h_bkg.GetMaximum())
    h_sig.SetMaximum(ymax)

    c = ROOT.TCanvas(f"c_{var}", var, 800, 600)

    h_sig.Draw("HIST")
    h_bkg.Draw("HIST SAME")

    leg = ROOT.TLegend(0.65, 0.7, 0.88, 0.88)
    leg.AddEntry(h_sig, "Signal (MC)", "l")
    leg.AddEntry(h_bkg, "Background (data)", "l")
    leg.Draw()

    c.SaveAs(f"VM_{var}_{tag}_sig_vs_bkg.png")
```

```python
for var, cfg in var_cfg.items():
    plot_sig_vs_bkg(
        df_sig=MCMumVMtag_sig,
        df_bkg=dataMumVMtag_bkg,
        var=var,
        cfg=var_cfg[var],
        tag="Down_mum_tag"
    )
```
